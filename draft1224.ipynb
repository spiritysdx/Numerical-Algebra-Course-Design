```python
columns = list(data2.columns[1:])
columns
```

```python
# 分离x y数据构造字典
dict_columns = {}
for i in columns: 
    y = data2.loc[:,i]
    temp = []
    for j in columns:
        if j != i:
            temp.append(j)
    m = data2.loc[:, temp[0]]
    tp = [m]
    for j in temp[1:]:
        tp.append(data2.loc[:, j])
    x = pd.DataFrame(tp).T
    data = {
        "x": x,
        "y": y
    }
    dict_columns[i] = data
```

```python
len(dict_columns)
```

```python
# 运算构造参数W字典 
# W = (A转置×A)的逆×A转置×b
dict_W = {}
for elem in dict_columns:
    A = np.array(dict_columns[elem]["x"].values)
    b = np.array(dict_columns[elem]["y"].values)
    A1 = np.linalg.inv(np.dot(A.T,A))
    W1 = np.dot(A1,A.T)
    W = np.dot(W1,b)
    dict_W[elem] = W
```

```python
# 残差字典 
# r = max(|A×W - b|)
dict_r = {}
for elem,w in zip(dict_columns, dict_W):
    A = np.array(dict_columns[elem]["x"].values)
    b = np.array(dict_columns[elem]["y"].values)
    W = dict_W[w]
    temp1 = np.dot(A,W) - b
    temp2 = abs(temp1)
    temp3 = max(temp2)
    dict_r[w] = temp3
dict_r
```
